```r
horas <- c(10, 20, 30, 40, 50, 60, 70, 80) # Horas de desenvolvimento
complexidade <- c(5, 3, 6, 7, 2, 8, 7, 4) # Complexidade do código
bugs <- c(2, 5, 7, 9, 5, 11, 14, 15) # Número de bugs

# Criando um DataFrame
dados = data.frame(horas, complexidade, bugs)

modelo_simples = lm(bugs ~ horas, data = dados)
modelo_multiplo = lm(bugs ~ horas + complexidade, data = dados)

summary(modelo_simples)
summary(modelo_multiplo)
```

```r
# Verificando Linearidade

plot(dados$horas, dados$bugs)
abline(modelo_simples, col = "red")
```

```r
# Verificando Homoscedasticidade

par(mfrow=c(1,2))
plot(modelo_simples$fitted.values, resid(modelo_simples))
abline(h = 0, col = "red")

plot(modelo_multiplo$fitted.values, resid(modelo_multiplo))
abline(h = 0, col = "red")
```

```r
library(car)

# Verificação de Normalidade (Gráfico Q-Q)
par(mfrow=c(1,2))
qqPlot(modelo_simples, main = "Modelo Simples")

qqPlot(modelo_multiplo, main = "Modelo Múltiplo")

# Em resumo, o gráfico Q-Q é uma ferramenta diagnóstica fundamental para verificar a normalidade dos resíduos em um modelo de regressão linear.
# Avaliar corretamente a distribuição dos resíduos é essencial para garantir a precisão e confiabilidade das inferências estatísticas feitas a partir do modelo.
# Se a normalidade for violada, pode ser necessário considerar transformações de dados ou modelos estatísticos alternativos.
```
